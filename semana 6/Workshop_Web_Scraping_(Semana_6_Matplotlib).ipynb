# Matplotlib (https://matplotlib.org/)
Matplotlib é uma biblioteca de plotagem poderosa em Python usada para criar visualizações estáticas, animadas e interativas. O objetivo principal do Matplotlib é fornecer aos usuários ferramentas e funcionalidades para representar dados graficamente, facilitando a análise e a compreensão.

**Diferentes tipos de gráficos no Matplotlib**

Matplotlib oferece uma ampla variedade de tipos de gráficos para atender a diversas necessidades de visualização de dados. Aqui estão alguns dos tipos de gráficos mais comumente usados ​​no Matplotlib:

* Gráfico de linha
* Gráfico de coluna
* Gráfico de barras
* Histogramas
* Gráfico de dispersão
* Gráfico de pilha
* Gráfico de caixa
* Gráfico de pizza
* Gráfico de erro
* Gráfico de violino
* Gráficos 3D

Sintaxe: matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)

Parâmetros:

Esta função aceita parâmetros que nos permitem definir escalas de eixos e formatar os gráficos. Esses parâmetros são mencionados abaixo:

* plot(x, y): plota x e y usando estilo de linha e cor padrão.
* plot.axis([xmin, xmax, ymin, ymax]): dimensiona o eixo x e y do valor mínimo ao máximo
* plot(x, y, color=’green’, marker=’o’, linestyle=’dashed’, linewidth=2, markersize=12): as coordenadas x e y são marcadas usando marcadores circulares de tamanho 12 e linha de cor verde com estilo de largura 2
* plot.xlabel('Eixo X'): nomeia o eixo x
* plot.ylabel('Eixo Y'): nomeia o eixo y
* plot(x, y, label = ‘Linha de amostra’): a linha de amostra plotada será exibida como uma legenda

```python
import matplotlib.pyplot as plt # pip install matplotlib
```

```python
# exemplo de plot simples de tipo linha
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
# para outro tipo de grficos, chamamos o metodo correspondente, ex.:
#  plt.bar() # para grafico de barra
# plt.hist() # para histograma

plt.axis([0, 6, 0, 20])

plt.show()
```

```python
# Eixo x contém os anos (year) e
# brazil_power_consumption & argentina_power_consumption são os valores do eixo y
year = [1972, 1982, 1992, 2002, 2012]
brazil_power_consumption = [100.6, 158.61, 305.54, 394.96, 724.79]
argentina_power_consumption = [10.5, 25.21, 58.65, 119.27, 274.87]

# plotando eixo x o ano (year) e
# eixo y consumo de energia (power consumption)
# com rótulos coloridos para cada país

plt.plot(year, brazil_power_consumption, color ='orange',
         label ='Brasil')

plt.plot(year, argentina_power_consumption, color ='green',
         label ='Argentina')

# nomeando eixos x e y
plt.xlabel('Anos')
plt.ylabel('Consumo de energia em kWh')

# nomeando o titulo do plot
plt.title('Consumo de energia per capita \
de Brasil e Argentina')

plt.legend()
plt.show()
```

Utlizando as bibliotecas pandas e matplotlib, e usando os dados de: https://raw.githubusercontent.com/fmassaretto/Workshop-WebScraping/main/arquivos/IDH.csv

Crie um grafico de barras mostrando os idhs de longevidade, renda e educação

```python
import pandas as pd
import matplotlib.pyplot as plt

# Pegando os dados de IDH: Renda, educação e Longevidade com Pandas
idh_data = pd.read_csv('https://raw.githubusercontent.com/fmassaretto/Workshop-WebScraping/main/arquivos/IDH.csv')
idh_location = idh_data['Territorialidade']
idh_income = idh_data['IDHM Renda']
idh_education = idh_data['IDHM Educação']
idh_longevity = idh_data['IDHM Longevidade']

# montar o grafico de barras
plt.bar(idh_location, idh_longevity, label="IDH Longevidade")
plt.bar(idh_location, idh_income, label="IDH Renda")
plt.bar(idh_location, idh_education, label="IDH Educação")

plt.xticks(rotation='vertical') # para que os nomes em x fiquem na vertical e fique legível

plt.legend(loc='center left') # loc, move a legenda no grafico, conforme as opções abaixo:
                              # 'best' (padrão se nao especificar)
                              # 'upper right'
                              # 'upper left'
                              # 'lower left'
                              # 'lower right'
                              # 'right'
                              # 'center left'
                              # 'center right'
                              # 'lower center'
                              # 'upper center'
                              # 'center'
plt.show()
```

Exemplo do grafico de barras na horizontal

```python
import matplotlib.pyplot as plt
import numpy as np

# Estado randômico fixo para reprodutividade constante
np.random.seed(19680801)

fig, ax = plt.subplots() # Com a função subplot() você pode desenhar vários gráficos em uma figura

# Exemplo de dados
people = ('Tom', 'Steve', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people)) # Exemplo fictício de erro do dados para mais e para menos, depois ele é definido na linha de baixo (xerr=)

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # Os rótulos lidos de cima para baixo
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```
